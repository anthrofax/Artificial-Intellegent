```python
# Import library yang diperlukan
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Fungsi keanggotaan Trapesium
def trapezoid(x, a, b, c, d):
    return fuzz.trapmf(x, [a, b, c, d])

# Fungsi keanggotaan Gaussian
def gaussian(x, mean, sigma):
    return fuzz.gaussmf(x, mean, sigma)

# Range untuk atribut usia (Age)
x_age = np.arange(20, 61, 1)

# Definisikan parameter fungsi keanggotaan Trapesium untuk usia
# Dalam format (a, b, c, d), di mana a <= b <= c <= d
a_age_trapezoid = 20
b_age_trapezoid = 25
c_age_trapezoid = 45
d_age_trapezoid = 60

# Definisikan parameter fungsi keanggotaan Gaussian untuk usia
age_mean_gaussian = 40
age_sigma_gaussian = 10

# Definisikan fungsi keanggotaan Trapesium untuk usia
age_membership_trapezoid = trapezoid(x_age, a_age_trapezoid, b_age_trapezoid, c_age_trapezoid, d_age_trapezoid)

# Definisikan fungsi keanggotaan Gaussian untuk usia
age_membership_gaussian = gaussian(x_age, age_mean_gaussian, age_sigma_gaussian)

# Plot fungsi keanggotaan Trapesium dan Gaussian untuk usia
plt.figure(figsize=(8, 5))
plt.plot(x_age, age_membership_trapezoid, label='Trapezoid')
plt.plot(x_age, age_membership_gaussian, label='Gaussian')
plt.title('Membership Functions for Age')
plt.xlabel('Age')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
plt.show()
```
